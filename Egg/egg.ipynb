```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyvista as pv
```

```python
perm = np.load('egg_perm.npy')
well = np.load('egg_well.npy')
mask = np.load('egg_mask.npy')
print('Perm: {} | Wells: {} | Mask: {}'.format(perm.shape, well.shape, mask.shape))

permx = perm[...,0]
print('PermX: {}'.format(permx.shape))
```

```python
p = pv.Plotter(shape=(3,12))
k = 0
for i in range(3):
    for j in range(12):
        p.subplot(i,j)
        p.add_mesh(permx[k]*mask, cmap='jet')
        k += 1
p.show(jupyter_backend='static', window_size=(1500,500))
```
